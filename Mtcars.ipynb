```python
# import library yang dibutuhkan
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load data mtcars
datas = pd.read_csv('data-mtcars.csv');
datas.head()
```

```python
# ukuran data (baris,kolom)
datas.shape
```

```python
# menambahkan kolom mpg_level
datas['mpg_level'] = datas.apply(lambda row : 'low' if (row['mpg']<20) else ( 'medium' if (row['mpg']>=20 and row['mpg']<=30) else 'high'), axis=1)
datas.head(20)
```

```python
# informasi data dari setiap atribut
datas.info()
```

```python
# perhitungan statististika dasar
datas.describe()
```

```python
# korelasi antar atribut
corr = datas.corr()
fig, ax = pyplot.subplots(figsize=(8,8))
sns.heatmap(corr, annot=True, fmt='.1f', ax=ax)
```

```python
# frekuensi setiap atribut berdasarkan kelompok mpg_level
f, ax = plt.subplots(figsize=(12,12))
datas.groupby('mpg_level').sum().unstack(0).plot.barh()
```

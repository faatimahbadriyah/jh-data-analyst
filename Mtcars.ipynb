```python
import pandas as pd
```

```python
datas = pd.read_csv('data-mtcars.csv');
```

```python
datas.head()
```

```python
# adding mpg_columns
datas_level = datas
datas_level['mpg_level'] = datas_level.apply(lambda row : 'low' if (row['mpg']<20) else ( 'medium' if (row['mpg']>=20 and row['mpg']<=30) else 'high'), axis=1)
```
